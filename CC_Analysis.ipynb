## UNCOVERING CUSTOMERS CHURNS

```python
plt.figure(figsize=(4, 4))
gb = df.groupby('Churn').agg({'Churn':"count"})
gb
plt.pie(gb['Churn'],labels= gb.index,autopct="%1.2f%%")
plt.title("Percentage of Churned customers",fontsize=10)
plt.show()
```

```python
# we counted the no of male and female along with no of churned females and males
plt.figure(figsize=(4, 4))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(4, 4))
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
plt.figure(figsize=(4, 4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
counts = df.groupby(["SeniorCitizen", "Churn"]).size().reset_index(name="count")
totals = counts.groupby("SeniorCitizen")["count"].transform("sum")
counts["percent"] = counts["count"] / totals * 100

pivot = counts.pivot(index="SeniorCitizen", columns="Churn", values="percent")

ax = pivot.plot(kind="bar", stacked=True, figsize=(5,5), colormap="cividis")

for i, row in pivot.iterrows():
    cumulative = 0
    for churn, value in row.items():
        if not pd.isna(value):
            plt.text(i, cumulative + value/2, f"{value:.1f}%",
                     ha="center", va="center", color="white", fontsize=10)
            cumulative += value

plt.title("Churn % by SeniorCitizen")
plt.ylabel("Percentage")
plt.xlabel("SeniorCitizen (0 = No, 1 = Yes)")
plt.legend(title="Churn")
plt.show()
```

```python
# comparatively a greater percentage people in senior citizen category have churned
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define subplot grid (3 rows × 3 columns since you have 9 features)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()  # flatten 2D array of axes into 1D for easy looping

# Loop through columns and create countplots
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, hue="Churn", ax=axes[i])
    axes[i].set_title(f"{col} by Churn")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

# Hide any unused axes (in case cols < grid size)
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
